```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
from scipy.ndimage import gaussian_filter1d

# Distance and energy data
distance = np.array([4, 5, 6, 7, 8, 9, 10])
energy = np.array([
    -1045.47236207245,
    -1045.47276299115,
    -1045.4727280125,
    -1045.47269356027,
    -1045.47267401961,
    -1045.47266271789,
    -1045.47265596726
])

# Interpolate data for smoother curve
distance_new = np.linspace(distance.min(), distance.max(), 300)  # 300 points for smooth curve
spl = make_interp_spline(distance, energy, k=3)  # Cubic spline interpolation
energy_smooth = spl(distance_new)

# Apply Gaussian smoothing
energy_smooth = gaussian_filter1d(energy_smooth, sigma=2)

# Plot the smoothed energy vs. distance
plt.figure(figsize=(10, 6))
plt.plot(distance_new, energy_smooth, color='b')

# Find the index of the minimum energy
min_energy_idx = np.argmin(energy_smooth)
min_energy = energy_smooth[min_energy_idx]
min_distance = distance_new[min_energy_idx]

# Add labels and title
plt.xlabel('Distance (Ã…)')
plt.ylabel('Energy')
plt.title('Energy vs. Distance')

# Show the plot
plt.show()
```
